## numpyÎ°ú Îç∞Ïù¥ÌÑ∞ ÏÉùÏÑ±Ìï¥ÏÑú ÏãúÍ∞ÅÌôîÌï¥Î≥¥Ïûê.
> numpyÎ•º ÏÉùÏÑ±ÌïòÍ∏∞ ÏúÑÌï¥ÏÑúÎäî `np.arange()`Î•º ÏÇ¨Ïö©ÌïúÎã§. <br>
> Ïôú `numpy`Î•º ÏÇ¨Ïö©Ìï¥ÏïºÌïòÎäîÏßÄ ÏãúÍ∞Ñ Ìö®Ïú® Ï∏°Î©¥ÏóêÏÑú Í≤∞Í≥ºÎ•º ÌôïÏù∏Ìï¥Î≥¥Ïûê.

```python
# NumPyÎ•º ÏÇ¨Ïö©ÌïòÏßÄ ÏïäÍ≥† pythonÏùò range() Ìï®ÏàòÎ°ú ÏÉùÏÑ±ÌñàÏùÑ Í≤ΩÏö∞ ÏÜçÎèÑ
%%time
result = []
for i in range(1000):
    result.append(i)
```

```python
# NumPyÎ•º ÏÇ¨Ïö©ÌñàÏùÑ Í≤ΩÏö∞ ÏÜçÎèÑ -> ÎÜÄÎùºÏö¥ ÏãúÍ∞Ñ Ï∞®Ïù¥!
%%time
np.arange(1000)
```

```python
# xÏ∂ï Î∞∞Ïó¥ÏùÑ Î≥ÑÎèÑÎ°ú ÏßÄÏ†ï
x = np.array([1, 2, 3, 4, 5])
y = x**2

plt.plot(x, y, 'ro')  
```

```python
# Í∑∏Î¶º ÌÅ¨Í∏∞, ÏÑ† Ïä§ÌÉÄÏùº, ÏÉâÏÉÅ ÏßÄÏ†ï
plt.figure(figsize=(15, 5))

x = np.linspace(-10, 10, 100)

plt.plot(x, np.cos(x), color='green', linestyle='dashed')

plt.show()
```

```python
# xÏ∂ï, yÏ∂ï Î≤îÏúÑ ÏßÄÏ†ï

plt.figure(figsize=(15, 5))

plt.plot(x, np.cos(x), color='green', linestyle='dashed')

plt.xlim(-12, 15)
plt.ylim(-1.5, 1.5);
```

```python
# Í∑∏ÎûòÌîÑÍ∞Ä ÌôîÎ©¥Ïóê ÍΩâ Ï∞®Í≤å ÌëúÏãú
plt.plot(x, np.cos(x), color='green', linestyle='dashed')

plt.axis('tight')
```

```python
# xÏ∂ï, yÏ∂ï ÎπÑÏú®Ïù¥ ÎèôÏùºÌïòÍ≤å ÌëúÏãú
plt.plot(x, np.cos(x), color='green', linestyle='dashed')
plt.axis('equal')
```

```python
# Ï†úÎ™© ÌëúÏãú

plt.plot(x, np.cos(x), color='green', linestyle='dashed')

plt.title("ÏΩîÏÇ¨Ïù∏ Ìï®Ïàò")

plt.xlabel("x")
plt.ylabel("cos(x)")

plt.show()
```

```python
# Î≤îÎ°Ä ÌëúÏãú

plt.plot(x, np.cos(x), 'g-', label='cos(x)')
plt.plot(x, np.sin(x), 'r--', label='sin(x)')

plt.legend()

plt.show()
```
