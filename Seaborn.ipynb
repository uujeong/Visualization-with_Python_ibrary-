# Seaborn 시각화

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train = pd.read_csv("./data/car_insurance.csv")
train = train.iloc[:10000]
```

## Bar Plot

```python
# barplot 1
plt.figure(figsize=(15, 5))

sns.barplot(x=train['Region_Code'], y=train['Age'])

plt.title("Age Distribution")
plt.xticks(rotation=90)
plt.show()
```

```python
# 데이터 정렬

sorted_index = train.loc[:, ['Region_Code', 'Age']].groupby('Region_Code').mean().sort_values(by='Age', ascending=False).index
sorted_index
```

```python
# barplot 2
plt.figure(figsize=(15, 5))

sns.barplot(x='Region_Code', y='Age', data=train, order=sorted_index)

plt.xticks(rotation=90)
plt.title("지역별 연령 분포")
plt.show()
```

## Count Plot

```python
sns.countplot(x=train['Gender'])
plt.show()
```

```python
sns.countplot(x='Gender', data=train)
plt.show()
```

```python
sns.countplot(y='Gender', data=train, order=['Male', 'Female'])
```

```python
sns.countplot(x='Gender', hue='Previously_Insured', data=train)  
plt.show()
```

## Distribution Plot

```python
sns.displot(x='Age', data=train)
plt.show()
```

```python
sns.displot(x='Age', kind='kde', data=train)
plt.show()
```

```python
sns.displot(x='Gender', data=train)
plt.show()
```

```python
sns.displot(x='Gender', hue="Response", data=train)
plt.show()
```

## Box Plot

```python
sns.boxplot(x='Gender', y='Age', hue="Response", data=train, palette="PRGn")

plt.legend(loc="upper right")
plt.show()
```

## Violin Plot

```python
sns.violinplot(x='Gender', y='Age', hue="Response", data=train, palette="PRGn")
plt.legend(loc="upper right")
plt.show()
```

## Strip Plot

```python
sns.stripplot(x='Gender', y='Age', hue="Response", data=train)
plt.legend(loc="upper right")
plt.show()
```

## Swarm Plot

```python
%%time 
sns.swarmplot(x='Gender', y='Age', hue="Response", data=train)
plt.legend(loc="upper right")
plt.show()
```

## Scatter Plot

```python
sns.scatterplot(x='Age', y='Annual_Premium', data=train)
plt.show()
```

## LM Plot

```python
sns.lmplot(x='Age', y='Annual_Premium', data=train)
plt.show()
```

## KDE Plot

```python
sns.kdeplot(x='Age', y='Annual_Premium', data=train, shade=False)
plt.show()
```

## Joint Plot

```python
sns.jointplot(x='Age', y='Annual_Premium', data=train)
plt.show()
```

```python
sns.jointplot(x='Age', y='Annual_Premium', data=train, kind='kde')
plt.show()
```
